```python
import os
import matplotlib.pyplot as plt
from glob import glob
```

```python
def plot_pair(image_path, label_path):
    "이미지와 라벨을 비교합시다."
    row=len(os.listdir(image_path))
    
    
    
    image_paths=glob(os.path.join(image_path, '*'))
    label_paths=glob(os.path.join(label_path, '*'))
    
    image_paths.sort()
    label_paths.sort()
    
    plt.figure(figsize=(12, row*2))
    for idx in range(row):
        
        plt.subplot(round(row/2), 4, 2*idx+1)
        plt.title(f"Input({image_paths[idx].split('/')[-1]})")
        plt.imshow(plt.imread(image_paths[idx]))
        plt.axis('off')
        plt.subplot(round(row/2), 4, 2*idx+2)
        plt.title('Output')
        plt.imshow(plt.imread(label_paths[idx]))
        plt.axis('off')
    plt.tight_layout()
    
```

```python
os.listdir('results/Flickr/train_50/images')
```

```python
label_path='results/Flickr/train_50/images/input_label'
image_path='results/Flickr/train_50/images/synthesized_image'
```

```python
plot_pair(label_path, image_path)
```

```python
label_path='results/Flickr/train_300/images/input_label'
image_path='results/Flickr/train_300/images/synthesized_image'
```

```python
plot_pair(label_path, image_path)
```
